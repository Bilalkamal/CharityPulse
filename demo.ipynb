```python
import pandas as pd
import random

# Generating mock data for the CSV file

# Helper function to generate random EINs
def generate_ein():
    return f"{random.randint(10, 99)}-{random.randint(1000000, 9999999)}"

# Helper function to generate random names
def generate_name():
    first_names = ["Alex", "Jamie", "Sam", "Taylor", "Jordan", "Casey", "Riley", "Morgan"]
    last_names = ["Smith", "Johnson", "Williams", "Brown", "Jones", "Garcia", "Miller", "Davis"]
    return f"{random.choice(first_names)} {random.choice(last_names)}"

# Helper function to generate random salaries
def generate_salary():
    return f"${random.randint(30, 200)}K"

# Helper function to generate random social media handles
def generate_social_handle(platform):
    names = ["charity", "help", "aid", "hope", "support", "care", "fund"]
    return f"@{random.choice(names)}_{platform}{random.randint(1, 999)}"

# Sample data
data = {
    "EIN": [generate_ein() for _ in range(20)],
    "Charity Name": [f"Charity {i}" for i in range(1, 21)],
    "Charity Type": random.choices(["Education", "Health", "Environment", "Animal Welfare"], k=20),
    "Charity Form Submitted": random.choices(["Form 990", "Form 990-EZ", "Form 990-N"], k=20),
    "Main Contact Name": [generate_name() for _ in range(20)],
    "Main Contact Title": random.choices(["CEO", "Director", "Manager", "Coordinator"], k=20),
    "Main Contact Salary": [generate_salary() for _ in range(20)],
    "Ruling Year": random.choices(range(1980, 2023), k=20),
    "Charity Location": [f"{i} Main St, City{i}, State, Zip{i}" for i in range(1, 21)],
    "Charity Website": [f"www.charity{i}.org" for i in range(1, 21)],
    "Mission Statement": [f"Our mission is to support {random.choice(['education', 'health', 'environment', 'animal welfare'])} initiatives." for _ in range(20)],
    "Social Media Handles - Facebook": [generate_social_handle("FB") for _ in range(20)],
    "Social Media Handles - Twitter": [generate_social_handle("TW") for _ in range(20)],
    "Social Media Handles - Instagram": [generate_social_handle("IG") for _ in range(20)],
    "Number of Beneficiaries": [f"{random.randint(100, 10000)} beneficiaries" for _ in range(20)]
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Displaying the first few rows of the DataFrame to ensure it's correct
df.head()
```

```python
df.to_csv("charity_data.csv", index=False)
```
